```python
dmg1=ds1g[["Mouse ID","Timepoint", "Sex"]]
dmg1=dmg1.rename(columns={"Sex":"Count"})

dmg1a=dmg1.groupby(["Mouse ID", "Timepoint"]).count().reset_index()
dmg1b=dmg1a.loc[dmg1a["Count"] > 1] 
duplicate=dmg1b["Mouse ID"].unique()
duplicate
```

```python
duplicate=duplicate.astype(str)
#duplicate=['a203', 'g989', 'z969']
#dmg1a=ds1g
if len(duplicate)==1:
    dmg1c=ds1g.loc[ds1g["Mouse ID"] == duplicate[0]] 
    duplicateindex=pd.DataFrame(dmg1c.index.astype(int))
    dmg1d=ds1g.drop(duplicateindex[0])
    print(len(dmg1d["Mouse ID"].value_counts()))
else:
    for x in range(len(duplicate)):
        dmg1c=ds1g.loc[ds1g["Mouse ID"] == duplicate[x]]
        duplicateindex=pd.DataFrame(dmg1c.index.astype(int))
        dmg1d=ds1g.drop(duplicateindex[0])
        ds1g=dmg1d
        print(len(dmg1d["Mouse ID"].value_counts()))
dmg1d.sort_values("Timepoint").head()
```

```python
dmg2=pd.DataFrame(dmg1d[['Drug Regimen', 'Tumor Volume (mm3)']])
dmg2.to_csv('dmg2a.csv', index=True, header=True)
dmg2a=pd.read_csv('dmg2a.csv')

del dmg2a["Unnamed: 0"]
Tumor_mean=dmg2a.groupby(["Drug Regimen"]).mean()
Tumor_mean=Tumor_mean.rename(columns={"Tumor Volume (mm3)":"Mean Tumor Volume"})
Tumor_median=dmg2a.groupby(["Drug Regimen"]).median()
Tumor_median=Tumor_median.rename(columns={"Tumor Volume (mm3)":"Median Tumor Volume"})
Tumor_var=dmg2a.groupby(["Drug Regimen"]).var()
Tumor_var=Tumor_var.rename(columns={"Tumor Volume (mm3)":"Tumor Volume Variance"})
Tumor_std=dmg2a.groupby(["Drug Regimen"]).std()
Tumor_std=Tumor_std.rename(columns={"Tumor Volume (mm3)":"Tumor Volume Std. Dev."})
Tumor_se=dmg2a.groupby(["Drug Regimen"]).sem()
Tumor_se=Tumor_se.rename(columns={"Tumor Volume (mm3)":"Tumor Volume Std. Err."})

merge_TumorVol=pd.merge(Tumor_mean, Tumor_median, on=['Drug Regimen'])
merge_TumorVol=pd.merge(merge_TumorVol, Tumor_var, on=['Drug Regimen'])
merge_TumorVol=pd.merge(merge_TumorVol, Tumor_std, on=['Drug Regimen'])
merge_TumorVol=pd.merge(merge_TumorVol, Tumor_se, on=['Drug Regimen'])

merge_TumorVol["Mean Tumor Volume"]=merge_TumorVol["Mean Tumor Volume"].map("{:,.6f}".format)
merge_TumorVol["Median Tumor Volume"]=merge_TumorVol["Median Tumor Volume"].map("{:,.6f}".format)
merge_TumorVol["Tumor Volume Variance"]=merge_TumorVol["Tumor Volume Variance"].map("{:,.6f}".format)
merge_TumorVol["Tumor Volume Std. Dev."]=merge_TumorVol["Tumor Volume Std. Dev."].map("{:,.6f}".format)
merge_TumorVol["Tumor Volume Std. Err."]=merge_TumorVol["Tumor Volume Std. Err."].map("{:,.6f}".format)
merge_TumorVol
```

```python
grp_TumorVol = dmg2a.groupby(["Drug Regimen"]).agg({'Tumor Volume (mm3)': ['mean', 'median', 'var', 'std', 'sem']})
grp_TumorVol
```

```python
dmg5=pd.DataFrame(dmg1d[["Drug Regimen", "Mouse ID","Timepoint","Tumor Volume (mm3)"]])
dmg5.to_csv('dmg5a.csv', index=True, header=True)
dmg5a=pd.read_csv('dmg5a.csv')
del dmg5a["Unnamed: 0"]

treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
tumor_vol_list = []
sorted_dmg5=dmg1d.sort_values(["Drug Regimen", "Mouse ID","Timepoint"], ascending=True)
```

```python
Capomulin=[]
Ramicane=[]
Infubinol=[]
Ceftamin=[]


gs_dmg5=sorted_dmg5[['Drug Regimen', 'Mouse ID','Timepoint']]
gs_dmg5=gs_dmg5.groupby(["Drug Regimen", "Mouse ID"]).max()
gs_dmg5.to_csv('gs_dmg5.csv', index=True, header=True)
gs_dmg5a=pd.read_csv('gs_dmg5.csv')

Capo = gs_dmg5a.loc[gs_dmg5a["Drug Regimen"] == "Capomulin"].reset_index()
Rami = gs_dmg5a.loc[gs_dmg5a["Drug Regimen"] == "Ramicane"].reset_index()
Infu = gs_dmg5a.loc[gs_dmg5a["Drug Regimen"] == "Infubinol"].reset_index()
Ceft = gs_dmg5a.loc[gs_dmg5a["Drug Regimen"] == "Ceftamin"].reset_index()

for a in range(len(Capo['Mouse ID'])):
    Capo1=sorted_dmg5.loc[(sorted_dmg5["Mouse ID"]==Capo['Mouse ID'][a]) & (sorted_dmg5["Timepoint"]==Capo['Timepoint'][a])]
    Capo1a=sorted_dmg5.loc[(sorted_dmg5["Mouse ID"]==Capo['Mouse ID'][a])]
    Capo2=pd.DataFrame(Capo1)
    Capo2.to_csv('Capo2.csv', index=True, header=True)
    Capo3=pd.read_csv('Capo2.csv')
    Capomulin.append(Capo3['Tumor Volume (mm3)'][0])
    
for a in range(len(Rami['Mouse ID'])):
    Rami1=sorted_dmg5.loc[(sorted_dmg5["Mouse ID"]==Rami['Mouse ID'][a]) & (sorted_dmg5["Timepoint"]==Rami['Timepoint'][a])]
    Rami2=pd.DataFrame(Rami1)
    Rami2.to_csv('Rami2.csv', index=True, header=True)
    Rami3=pd.read_csv('Rami2.csv')
    Ramicane.append(Rami3['Tumor Volume (mm3)'][0])
    
for a in range(len(Infu['Mouse ID'])):
    Infu1=sorted_dmg5.loc[(sorted_dmg5["Mouse ID"]==Infu['Mouse ID'][a]) & (sorted_dmg5["Timepoint"]==Infu['Timepoint'][a])]
    Infu2=pd.DataFrame(Infu1)
    Infu2.to_csv('Infu2.csv', index=True, header=True)
    Infu3=pd.read_csv('Infu2.csv')
    Infubinol.append(Infu3['Tumor Volume (mm3)'][0])
    
for a in range(len(Ceft['Mouse ID'])):
    Ceft1=sorted_dmg5.loc[(sorted_dmg5["Mouse ID"]==Ceft['Mouse ID'][a]) & (sorted_dmg5["Timepoint"]==Ceft['Timepoint'][a])]
    Ceft2=pd.DataFrame(Ceft1)
    Ceft2.to_csv('Ceft2.csv', index=True, header=True)
    Ceft3=pd.read_csv('Ceft2.csv')
    Ceftamin.append(Ceft3['Tumor Volume (mm3)'][0])
```

```python
fig1, ax1 = plt.subplots()
ax1.boxplot([Capomulin, Ramicane, Infubinol, Ceftamin])
ax1.set_xlim([0,5])
ax1.set_xticklabels(treatment_list)
ax1.set_ylabel('Final Tumor Volume (mm3)')
plt.show()
```

```python
Q1=[]
Q3=[]
iqr=[]
B1=[]
B3=[]
IQRA=[]

IQ_Capo=[]
IQ_Rami=[]
IQ_Infu=[]
IQ_Ceft=[]

outlierinfu=[]

capo1=pd.DataFrame(Capomulin)
rami1=pd.DataFrame(Ramicane)
infu1=pd.DataFrame(Infubinol)
ceft1=pd.DataFrame(Ceftamin)
for b in treatment_list:
    if b == "Capomulin":
        quartiles=capo1.quantile([.25,.5,.75])
        Q1.append(quartiles[0][0.25])
        Q3.append(quartiles[0][0.75])
        iqr.append((quartiles[0][0.75]-quartiles[0][0.25]))
        B1.append(quartiles[0][0.25] - (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))
        B3.append(quartiles[0][0.75] + (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))    
        IQRA=[Q1, Q3, iqr, B1, B3]
        IQ_Capo.append(IQRA)
        IQ_Capo2=pd.DataFrame(IQ_Capo, columns=["Q1", "Q3", "IQR", "Lower Bound", "Upper Bound"])
        IQ_Capo2["Treatment"]="Capomulin"
        #IQ_Capo2["1"]="1"
        #IQ_Capo2.set_index(['Treatment'])
 
    elif b == "Ramicane":
        quartiles=rami1.quantile([.25,.5,.75])
        Q1.append(quartiles[0][0.25])
        Q3.append(quartiles[0][0.75])
        iqr.append((quartiles[0][0.75]-quartiles[0][0.25]))
        B1.append(quartiles[0][0.25] - (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))
        B3.append(quartiles[0][0.75] + (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))    
        IQRA=[Q1, Q3, iqr, B1, B3]
        IQ_Rami.append(IQRA)
        IQ_Rami2=pd.DataFrame(IQ_Rami, columns=["Q1", "Q3", "IQR", "Lower Bound", "Upper Bound"])
        IQ_Rami2["Treatment"]="Ramicane"
        #IQ_Rami2["1"]="1"
        #IQ_Rami2.set_index(['Treatment'])                

    elif b == "Infubinol":
        quartiles=infu1.quantile([.25,.5,.75])
        Q1.append(quartiles[0][0.25])
        Q3.append(quartiles[0][0.75])
        iqr.append((quartiles[0][0.75]-quartiles[0][0.25]))
        B1.append(quartiles[0][0.25] - (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))
        B3.append(quartiles[0][0.75] + (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))    
        IQRA=[Q1, Q3, iqr, B1, B3]
        IQ_Infu.append(IQRA)
        IQ_Infu2=pd.DataFrame(IQ_Infu, columns=["Q1", "Q3", "IQR", "Lower Bound", "Upper Bound"])
        IQ_Infu2["Treatment"]="Infubinol"
        #IQ_Infu2["1"]="1"
        #IQ_Infu2.set_index(['Treatment'])    

    elif b=='Ceftamin':
        quartiles=ceft1.quantile([.25,.5,.75])
        Q1.append(quartiles[0][0.25])
        Q3.append(quartiles[0][0.75])
        iqr.append((quartiles[0][0.75]-quartiles[0][0.25]))
        B1.append(quartiles[0][0.25] - (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))
        B3.append(quartiles[0][0.75] + (1.5*(quartiles[0][0.75]-quartiles[0][0.25])))    
        IQRA=[Q1, Q3, iqr, B1, B3]
        IQ_Ceft.append(IQRA)
        IQ_Ceft2=pd.DataFrame(IQ_Ceft, columns=["Q1", "Q3", "IQR", "Lower Bound", "Upper Bound"])
        IQ_Ceft2["Treatment"]="Ceftamin"
        #IQ_Ceft2["1"]="1"
        #IQ_Ceft2.set_index(['Treatment'])         
    Q1=[]
    Q3=[]
    iqr=[]
    B1=[]
    B3=[]
    IQRA=[] 
```

```python
Capomulinwt=[]
Capomulintv=[]
for a in range(len(Capo['Mouse ID'])):
    Capo1a=sorted_dmg5.loc[(sorted_dmg5["Mouse ID"]==Capo['Mouse ID'][a])]
    Capo2a=pd.DataFrame(Capo1a)
    Capo2a.to_csv('Capo2a.csv', index=True, header=True)
    Capo3a=pd.read_csv('Capo2a.csv')
    Capomulinwt.append(Capo3a['Weight (g)'][0])
    Capomulintv.append(Capo3a['Tumor Volume (mm3)'].mean())
    
plt.scatter(Capomulinwt, Capomulintv)
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```

```python
from scipy.stats import linregress
Capomulinwt1a = pd.DataFrame(Capomulinwt).astype(str)
Capomulintv1a = pd.DataFrame(Capomulintv).astype(str)
Capomulinwt1a.to_csv('Capowt2a.csv', index=True, header=True)
Capomulinwt1=pd.read_csv('Capowt2a.csv')

Capomulintv1a.to_csv('Capotv2a.csv', index=True, header=True)
Capomulintv1=pd.read_csv('Capotv2a.csv')
(slope, intercept, rvalue, pvalue, stderr) = linregress(Capomulinwt1["0"], Capomulintv1["0"])
rvalue1=round(rvalue,2)
rvalue1
```

```python
regress_values = Capomulinwt1["0"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
```

```python
plt.scatter(Capomulinwt1["0"], Capomulintv1["0"])
plt.plot(Capomulinwt1["0"],regress_values,"r-")
plt.annotate(line_eq,(20,1),fontsize=15,color="red")
plt.ylabel("Average Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
print(f"The correlation between mouse weight and the average tumor volume is: {rvalue1}")
plt.show()
```
